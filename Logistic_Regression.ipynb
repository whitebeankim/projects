## 1. Download and pre-process the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get the training and testing data
df_train = pd.read_csv('./hw1-2-data/train.csv')
df_test  = pd.read_csv('./hw1-2-data/test.csv')

df_train.head()
```

```python
# Show the passenger data variable info
df_train.info()
```

```python
# Merge the training and testing data
df_train_drop = df_train.drop(columns = ['Survived'])
df_all = pd.concat([df_train_drop, df_test], ignore_index=True)
train_num = df_train.shape[0]

df_all.head()
```

```python
# Remove the ID, Name, Ticket, and Cabin columns
df_all.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)
df_all.head()
```

```python
# Show the passenger data variable info
df_all.info()
```

```python
# Fill the age info for missing data as an average
age_avg = df_all['Age'].mean()
print(age_avg)
df_all['Age'] = df_all['Age'].fillna(age_avg)

# Fill the fare info for missing data as an average
fare_avg = df_all['Fare'].mean()
print(fare_avg)
df_all['Fare'] = df_all['Fare'].fillna(fare_avg)

# Fill the Embarked info for missing data as 'C'
df_all['Embarked'] = df_all['Embarked'].fillna('C').astype(str)
df_all.head()
```

```python
# Show the passenger data variable info
df_all.info()
```

```python
# Convert male to 1 and female to 0 in 'Sex' column
df_all['Sex'] = df_all['Sex'].map({'male': 1, 'female': 0})

# Convert C to 2, Q to 1, and S to 0 in 'Embarked' column
df_all['Embarked'] = df_all['Embarked'].map({'C': 2, 'Q': 1, 'S': 0, 'NA': -1})

df_all.head()
```

```python
# Scale down the Age and Fare values by dividing by 10
df_all['Age'] = df_all['Age'] / 10
df_all['Fare'] = df_all['Fare'] / 10

df_all.head()
```

## 2. Implement logistic regression
Name, Ticket, Cabin: Removed   
Sex, Embarked: Converted to the integer   
Age, Fare: Scaled down (Divided by 10)   

```python
# Get the df_train data set
df_train_final = df_all.iloc[:train_num]

# Put feature variable to X
X = df_train_final

# Put response variable to y
y = df_train['Survived']
```

```python
# Split training data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
# Implement logistic regression
from sklearn.linear_model import LogisticRegression

# Instantiate the model
logreg = LogisticRegression(random_state=16, solver='liblinear', max_iter=200)

# Fit the model with data
logreg.fit(X_train, y_train)

# Calculate the predicted value from the test set
y_test_pred = logreg.predict(X_test)

# Evaluate predictions on the training set by calculating the correct rate
print("Correct rate: ", sum(y_test_pred == y_test)/len(y_test))
```

## 3. Test the classifier using the testing data

```python
# Extract the testing data from the pre-processed data
df_test_new = df_all.iloc[train_num:]

# Esitmate whether the passenger survived from the testing data
y_pred = logreg.predict(df_test_new)

# Make the prediction CSV file
df_predict = pd.DataFrame(y_pred).rename(columns={0:'Survived'})
df_predict['PassengerId'] = df_test['PassengerId']

df_predict.to_csv('./hw1-2-data/test_predict.csv', index=False)
```
